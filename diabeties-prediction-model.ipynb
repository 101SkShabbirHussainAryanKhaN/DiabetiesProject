```python
class GDRegressor:
    """
    Gradient Descent Linear Regressor.

    Attributes:
        coef_ (numpy array): Model coefficients.
        intercept_ (float): Model intercept.
        lr (float): Learning rate.
        epochs (int): Number of training epochs.
    """

    def __init__(self, learning_rate=0.01, epochs=100):
        """
        Initialize the regressor.

        Args:
            learning_rate (float, optional): Defaults to 0.01.
            epochs (int, optional): Defaults to 100.
        """
        self.coef_ = None
        self.intercept_ = None
        self.lr = learning_rate
        self.epochs = epochs

    def fit(self, X_train, y_train):
        """
        Train the model.

        Args:
            X_train (numpy array): Training features.
            y_train (numpy array): Training targets.
        """
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])

        for i in range(self.epochs):
            # Predict y values
            y_hat = np.dot(X_train, self.coef_) + self.intercept_

            # Calculate derivatives
            intercept_der = -2 * np.mean(y_train - y_hat)
            coef_der = -2 * np.dot((y_train - y_hat), X_train) / X_train.shape[0]
            

            # Update parameters
            self.coef_=self.coef_-(self.lr*coef_der)

           
        
            print(f'Epoch {i+1}: intercept = {self.intercept_}, coefficients = {self.coef_}')
      


    def predict(self, X_test):
        """
        Make predictions.

        Args:
            X_test (numpy array): Test features.

        Returns:
            numpy array: Predicted values.
        """
        return np.dot(X_test, self.coef_) + self.intercept_



```
